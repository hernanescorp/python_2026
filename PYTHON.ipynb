STRING AS VAR

```python
frase = 'Al que madruga Dios lo ayuda'
author = 'Hernan Escriba'
print('frase: ', frase, 'Autor: ', author)
```

Use of F-STRING

```python
frase = 'Pussies love me'
author = 'Hernan Escriba'

print(f'frase: {frase}, Autor: {author}')
```

Use of FORMAT

```python
print('Frase: {}, Autor: {}'.format(frase,author) )
```

Decimal format print:

```python
var = 3.141567
print('Valor: {:.2f}'.format(var))
```
